# case 1

```python
exp = "1/math.sqrt((0.04/(x*x)) + 0.31 + 0.65*x*x)"

i = 1.001
x_11 = []
y_11 = []
while i < 6:
    x_11.append(ml.Integral.simpson(exp, 1, i, 100))
    y_11.append(i)
    i += 0.001
```

```python
plt.plot(x_11, y_11)
```

```python
i = 0.999
x_12 = []
y_12 = []
while i > 0:
    x_12.append(ml.Integral.simpson(exp, 1, i, 100))
    y_12.append(i)
    i -= 0.001
```

```python
plt.plot(x_12, y_12)
```

```python
xvalues_1 = x_12 + x_11
yvalues_1 = y_12 + y_11
```

```python
plt.scatter(xvalues_1, yvalues_1, s = 0.5, label = r"$\Omega_{r,0} = 0.04$, $\Omega_{\Lambda, 0} = 0.65$")
```
