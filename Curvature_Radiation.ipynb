# case 1

```python
exp = "1/math.sqrt((0.69) + 0.31/(x*x))"
```

```python
i = 1.001
x_11 = []
y_11 = []
while i < 6:
    x_11.append(ml.Integral.simpson(exp, 1, i, 1000))
    y_11.append(i)
    i += 0.001

plt.plot(x_11, y_11)
```

```python
k = 0.999
x_12 = []
y_12 = []
while k > 0:
    x_12.append(ml.Integral.simpson(exp, 1, k, 1000))
    y_12.append(k)
    k -= 0.001

plt.plot(x_12, y_12)
```

```python
x_values_1 = x_12[::-1] + x_11
y_values_1 = y_12[::-1] + y_11
t_values_1 = [14.389*(value - x_values_1[0]) for value in x_values_1]

plt.plot(t_values_1, y_values_1)
```
