```python
import import_ipynb      #for importing other files in jupyter
import mylibrary_true_personal as ml   #for accessing library
import math             #for functions
import random as rand   #for random 
from matplotlib import pyplot as plt
```

# k = +1, one root

```python
exp = "1/math.sqrt((0.75/(x*x)) + (0.31/x) - 0.06)"

i = 1.001
x_11 = []
y_11 = []
while i < 6.962:
    x_11.append(ml.Integral.simpson(exp, 1, i, 100))
    y_11.append(i)
    i += 0.001
```

```python
i = 6.960
x_12 = []
y_12 = []
while i > 0:
    x_12.append(ml.Integral.simpson(exp, i, 6.961, 100) + x_11[len(x_11)-1])
    y_12.append(i)
    i -= 0.001
```

```python
i = 0.999
x_13 = []
y_13 = []
while i > 0:
    x_13.append(ml.Integral.simpson(exp, 1, i, 1000))
    y_13.append(i)
    i -= 0.001
```

```python
xvalues_1 = x_13[::-1] + x_11 + x_12
yvalues_1 = y_13[::-1] + y_11 + y_12
t_values_1 = [14.389*(value - xvalues_1[0]) for value in xvalues_1]

plt.plot(t_values_1, yvalues_1)
```

# k = -1, no root

```python
exp_2 = "1/math.sqrt((0.31/(x*x)) + (0.4/x) + 0.29)"

i = 1.001
x_21 = []
y_21 = []
while i < 10:
    x_21.append(ml.Integral.simpson(exp_2, 1, i, 1000))
    y_21.append(i)
    i += 0.001
```

```python
plt.plot(x_21, y_21)
```

```python
i = 0.999
x_22 = []
y_22 = []
while i > 0:
    x_22.append(ml.Integral.simpson(exp_2, 1, i, 1000))
    y_22.append(i)
    i -= 0.001
```

```python
plt.plot(x_22, y_22)
```

```python
xvalues_2 = x_22[::-1] + x_21
yvalues_2 = y_22[::-1] + y_21
t_values_2 = [14.389*(value - xvalues_2[0]) for value in xvalues_2]

plt.plot(t_values_2, yvalues_2)
```

```python
plt.figure(figsize=(10,7))
plt.grid()
plt.plot(xvalues_1, yvalues_1, label = "$\Omega_{r,0} = 0.75$, $\Omega_{m,0} = 0.31$, k = +1")
plt.plot(xvalues_2, yvalues_2, label = "$\Omega_{r,0} = 0.31$, $\Omega_{m,0} = 0.40$, k = -1")
plt.rc('font', size=12) 
plt.xlabel("$t$   (Billion years)")
plt.ylabel("a")
lgnd = plt.legend(loc="upper right", scatterpoints=1, fontsize=10)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
plt.show()
```
