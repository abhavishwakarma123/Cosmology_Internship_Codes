# k = -1, one root

```python
exp = "1/math.sqrt((0.75/(x*x)) + (0.31/x) - 0.06)"

i = 1.001
x_11 = []
y_11 = []
while i < 6.962:
    x_11.append(ml.Integral.simpson(exp, 1, i, 100))
    y_11.append(i)
    i += 0.001
```

```python
plt.scatter(x_11, y_11, s = 0.5)
```

```python
i = 6.960
x_12 = []
y_12 = []
while i > 0:
    x_12.append(ml.Integral.simpson(exp, i, 6.961, 100) + x_11[len(x_11)-1])
    y_12.append(i)
    i -= 0.001

plt.plot(x_12, y_12)
```

```python
i = 0.999
x_13 = []
y_13 = []
while i > 0:
    x_13.append(ml.Integral.simpson(exp, 1, i, 1000))
    y_13.append(i)
    i -= 0.001

plt.plot(x_13, y_13)
```

```python
xvalues_1 = x_13 + x_11 + x_12
yvalues_1 = y_13 + y_11 + y_12

plt.scatter(xvalues_1, yvalues_1, s = 0.5)
```
