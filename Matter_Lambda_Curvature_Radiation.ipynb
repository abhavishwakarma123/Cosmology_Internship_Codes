# +ve curvature, +ve lambda

```python
exp = "1/math.sqrt((0.00009/(x*x)) + (0.31/x) + 0.69*x*x - 0.00009)"

i = 1.001
x_11 = []
y_11 = []
while i < 4.5:
    x_11.append(ml.Integral.simpson(exp, 1, i, 1000))
    y_11.append(i)
    i += 0.001

plt.plot(x_11, y_11)
```

```python
i = 0.999
x_12 = []
y_12 = []
while i > 0:
    x_12.append(ml.Integral.simpson(exp, 1, i, 1000))
    y_12.append(i)
    i -= 0.001
```

```python
plt.plot(x_12, y_12)
```

```python
xvalues_1 = x_12[::-1] + x_11
yvalues_1 = y_12[::-1] + y_11
t_values_1 = [14.389*(value - xvalues_1[0]) for value in xvalues_1]
plt.plot(t_values_1, yvalues_1)
```
