# less radiation, positive lambda

```python
exp_1 = "1/math.sqrt((0.04/(x*x)) + (0.31/x) + 0.65*x*x)"

i = 1.001
x_11 = []
y_11 = []
while i < 3:
    x_11.append(ml.Integral.simpson(exp_1, 1, i, 1000))
    y_11.append(i)
    i += 0.001
```

```python
i = 0.999
x_12 = []
y_12 = []
while i > 0:
    x_12.append(ml.Integral.simpson(exp_1, 1, i, 1000))
    y_12.append(i)
    i -= 0.001
```

```python
xvalues_1 = x_12[::-1] + x_11
yvalues_1 = y_12[::-1] + y_11
t_values_1 = [14.389*(value - xvalues_1[0]) for value in xvalues_1]
```
