```python
import import_ipynb      #for importing other files in jupyter
import mylibrary_true_personal as ml   #for accessing library
import math             #for functions
import random as rand   #for random 
from matplotlib import pyplot as plt
import numpy as np
```

```python
exp = "1/math.sqrt((1.3/x) - 0.3)"

i = 1.001
x_11 = []
y_11 = []
while i < 4.332:
    x_11.append(ml.Integral.simpson(exp, 1, i, 1000))
    y_11.append(i)
    i += 0.001

plt.plot(x_11, y_11)
```

```python
i = 0.999
x_12 = []
y_12 = []
while i > 0:
    x_12.append(ml.Integral.simpson(exp, 1, i, 1000))
    y_12.append(i)
    i -= 0.001

plt.plot(x_12, y_12)
```

```python
i = 4.332
x_13 = []
y_13 = []
while i > 0:
    x_13.append(ml.Integral.simpson(exp, i, 4.332, 1000) + x_11[len(x_11)-1])
    y_13.append(i)
    i -= 0.001

plt.plot(x_13, y_13)
```

```python
x_values_1 = x_12[::-1] + x_11 + x_13
y_values_1 = y_12[::-1] + y_11 + y_13
t_values_1 = [14.389*(value - x_values_1[0]) for value in x_values_1]

plt.plot(t_values_1, y_values_1)
```

```python
exp_2 = "1/math.sqrt((0.31/x) + 0.69)"

i = 1.001
x_21 = []
y_21 = []
while i < 10:
    x_21.append(ml.Integral.simpson(exp_2, 1, i, 1000))
    y_21.append(i)
    i += 0.001

plt.plot(x_21, y_21)
```

```python
i = 0.999
x_22 = []
y_22 = []
while i > 0:
    x_22.append(ml.Integral.simpson(exp_2, 1, i, 1000))
    y_22.append(i)
    i -= 0.001

plt.plot(x_22, y_22)
```

```python
x_values_2 = x_22[::-1] + x_21
y_values_2 = y_22[::-1] + y_21
t_values_2 = [14.389*(value - x_values_2[0]) for value in x_values_2]

plt.plot(t_values_2, y_values_2)
```
