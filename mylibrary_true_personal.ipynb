```python
#####################################################################
'''This class performs AP, GP, HP sums'''
#####################################################################

class Summing:
    
    def ap(first = 1.0, n = 5.0, common_diff = 1.5):

        summ = first

        n_flor = math.floor(n)                   #find the floor value of the input n
        if n <= 0 or (n-n_flor) != 0:            #checks if the input is not a positive integer
            print("Invalid input")
        else:
            n = int(n)                           #because n is taken as a float input
            for x in range(n - 1):
                summ += common_diff              
            return summ
    
    def gp(first_GP = 1.0, n_GP = 5.0, common_ratio_GP = 0.5):
        
        sum_GP = first_GP
        add = first_GP
        n_GP_flor = math.floor(n_GP)                      #find the floor value of the input n

        if n_GP <= 0 or (n_GP-n_GP_flor) != 0:            #checks if the input is not a positive integer
            print("Invalid input")
        else:
            n_GP = int(n_GP)                              #because n_GP is taken as a float input
            for x in range(n_GP - 1):
                add = add * common_ratio_GP               #because if I directly iterate on sum 
                sum_GP += add                             #the common ratio will not be multiplied by the first value in further iterations
            return sum_GP
        
    def hp(first_HP = 1.0, n_HP = 5, common_diff_HP = 1.5):
        
        sum_HP = first_HP
        n_HP_flor = math.floor(n_HP)                      #find the floor value of the input n

        if n_HP <= 0 or (n_HP-n_HP_flor) != 0:            #checks if the input is not a positive integer
            print("Invalid input")
        else:
            add_HP = 1 / first_HP                         #to add the common difference
            n_HP = int(n_HP)
            for x in range(n_HP - 1):
                add_HP += common_diff_HP
                sum_HP += 1 / add_HP
            return sum_HP
        
    def sum_odd(k = 10):
        
        sum_new = 0                                         #sets the sum to zero

        k_flor = math.floor(k)
        if k <= 0 or (k-k_flor) != 0:                       #checks if the input is not a positive integer
            print("Invalid Input")
        else:
            k = int(k)                                      #converts the float value to int
            for i in range(2*k):
                if i%2 != 0:
                    sum_new+=i
                    
            return sum_new
        
```

```python
#####################################################################
'''This class is for the assignment 1 question'''
#####################################################################

class Assign_1:
    
    def factorial(n):

        n_flor = math.floor(n)

        if n < 0 or (n-n_flor) != 0:
            print("Invalid Input")
        else:
            n = int(n)
            fct = 1
            if n == 0:
                return fct
            else:
                for x in range(1, n+1):
                    fct = fct * x
                return fct
    
    def acc_sin(x, n = 50, epsilon = 0.00001):      #defining the accurate sin function to get accurate value of sin
        accrt_sin = 0
        actual_sin = math.sin(x)
        for i in range(1, n):
            if abs(actual_sin - accrt_sin) > epsilon:
                accrt_sin += ( pow(x, (2 * i - 1) ) * pow(-1, i+1) ) / Assign_1.factorial(2 * i - 1)
        return accrt_sin
    
    def acc_exp_inv(x, n = 50, epsilon = 0.00001):     #defining the accurate exponential function 
        accrt_exp = 0
        actual_exp = math.exp(-1 * x)
        for j in range(n):
            if abs(actual_exp - accrt_exp) > epsilon:
                accrt_exp += ( pow(-x, j) ) / Assign_1.factorial(j)
        return accrt_exp
    
    def sin_per_itertn(x, n):    #defining the sin function for certain number of iterations
        sin_val = 0
        for m in range(n+1):
            sin_val += ( pow(x, (2 * m + 1) ) * pow(-1, m) ) / Assign_1.factorial(2 * m + 1)
        return sin_val
    
    def exp_inv_per_itertn(x, n):   #defining the exp function for certain number of iterations
        exp_val = 1
        for p in range(1, n+1):
            exp_val += pow(-x, p) / Assign_1.factorial(p)
        return exp_val
    
    def plot_sin(x, epsilon = 0.00001):
        error = []
        iteration_number = []
        actual = math.sin(x)
        s = 1

        while abs(actual - Assign_1.sin_per_itertn(x, s)) > epsilon:
            iteration_number.append(s)
            error.append(abs(Assign_1.sin_per_itertn(x, s) - actual))
            s = s + 1

        plt.title("Absolute error vs iteration number for sin")
        plt.xlabel("Iteration numbers")
        plt.ylabel("Absolute error")
        plt.plot(iteration_number, error, '-h')
        plt.show()
        
    def plot_exp_inv(x, epsilon = 0.00001):
        error = []
        iteration_number = []
        actual = math.exp(-1 * x)
        e = 1
        
        while abs(actual - Assign_1.exp_inv_per_itertn(x, e)) > epsilon:
            error.append(abs(Assign_1.exp_inv_per_itertn(x, e) - actual))
            iteration_number.append(e)
            e = e + 1

        plt.title("Absolute error vs iteration number for exponenetial")
        plt.xlabel("Iteration numbers")
        plt.ylabel("Absolute error")
        plt.plot(iteration_number, error)
        plt.show()
```

```python
##################################################################
'''This class performs all row operations'''
##################################################################
class Row_operations:
    
    def swap_rows(matrix, n, r, col):                  #swapping rows using traditional method
        temp = 0
        for j in range(col):
            temp = matrix[n][j]
            matrix[n][j] = matrix[r][j]
            matrix[r][j] = temp
        return matrix
    
    def multiply_row(matrix, r = [], fact = 0):        #multiplying a factor to the row
        row = []
        col = len(matrix[0])
        for x in range(col):
            row.append(fact*r[x])
        return row
    
    def add_rows(matrix, r1, r2):                      #adding two rows
        row = []
        col = len(matrix[0])
        for x in range(col):
            row.append(r1[x] + r2[x])
        return row
    
    def partial_pivot(matrix, p):                      #partial pivoting
        pivot = matrix[p][p]
        row = len(matrix)
        col = len(matrix[0])
        if pivot != 0:
            return matrix
        else:
            for i in range(p+1,row):
                if matrix[i][p] != 0:
                    Row_operations.swap_rows(matrix, i, p, col)
                    return matrix                      #will return as soon as a non zero pivot is
                                                       #found
```

```python
##################################################################
'''This class performs all matrix operations'''
##################################################################
class Matrix_operations:
    
    def augment_matrix(M_square, M_col):                  #augmenting two matrices
        M = []
        
        row = len(M_square)
        col = len(M_square[0])
        
        for i in range(row):
            M.append(M_square[i])

        for i in range(row):
            M[i] = np.append(M[i], M_col[i])           #np.append because np.arrays do not
                                                          #work with normal append function
        return M
            
    def make_proper_column(M):                            #makes a list a column matrix
        C = []                                            #of the form [[],[],[]...]    
        for i in M:
            temp = []
            temp.append(i)
            C.append(temp)
        return C
            
    def make_proper_row(M):                               #makes a list a row matrix
        R = []                                            #of the form [...]
        R.append(M)
        return R
    
    def mat_mult(X, Y):                                   #DEFINING THE MATRIX MULTIPLICATION FUNCTION
        len_row_X = len(X)
        len_column_X = len(X[0])
        len_row_Y = len(Y)
        len_column_Y = len(Y[0])

        if len_column_X == len_row_Y:                     #CHECKING IF MATRIX MULTIPLICATION IS POSSIBLE
            Z = []                                        #THIS WILL BE FINAL MULTIPLIED MATRIX
            for i in range(len(X)):
                Z_row_temp = []                           #THIS WILL BE THE TEMPORARY ROWS APPENDED TO Z AFTER EACH ITERATION
                for j in range(len(Y[0])):
                    temp = 0                              #THIS WILL BE THE TEMPORARY ELEMENTS OF EACH ROW
                    for k in range(len(Y)):
                        temp += X[i][k]*Y[k][j]           #THE MAIN ITERATION
                    Z_row_temp.append(temp)               #APPENDING THE ELEMENT TO THE ROW
                Z.append(Z_row_temp)                      #APPENDING THE ROW TO Z
            return Z
        else:
            print("Invalid Argument")
            
    def separate_aug_mat(M):                              #separating the augmented matrix
        A = []                                            #to its counterparts
        B = []
        row = len(M)
        
        for i in range(row):
            temp_A_row = []
            temp_B_row = []

            for j in range(len(M[0])):
                if j < row:
                    temp_A_row.append(M[i][j])
                else:
                    temp_B_row.append(M[i][j])
            A.append(temp_A_row)
            B.append(temp_B_row)
            
        return A, B
    
    def generate_identity(n = 3):
    
        Z = []
        for i in range(n):
            temp = []
            for j in range(n):
                if i != j:
                    temp.append(0)
                else:
                    temp.append(1)
            Z.append(temp)
        return Z
    
    def transpose(X):
        n = len(X)

        trans = Matrix_operations.generate_identity(n)

        for i in range(n):
            for j in range(n):
                trans[i][j] = X[j][i]

        return trans
    
    def generate_null(n = 3):
    
        Z = []
        for i in range(n):
            temp = []
            for j in range(n):
                temp.append(0)
            Z.append(temp)
        return Z
                    
```

```python
##################################################################
'''This class performs rounding'''
##################################################################
class Round:                             

    def round_matrix(matrix, acc = 2):                                    #rounding each element of the matrix
        row = len(matrix)
        col = len(matrix[0])
        for i in range(row):
            for j in range(col):
                matrix[i][j] = round(matrix[i][j], acc)
        return matrix
    
    def round_num(num, acc = 2):                                          #rounding any number
        num_1 = num*pow(10, acc)
        if num >= 0:
            num_1 = math.floor(num_1)
        else:
            num_1 = math.ceil(num_1)
        num = num_1*pow(0.1, acc)
        return num
```

```python
##################################################################
'''This class performs all gauss jordan operations'''
##################################################################
class Gauss_jordan:
    
    def solve(matrix):                                             #doing the complete gauss jordan
        row = len(matrix)                                          #elimination
        col = len(matrix[0])

        
        
        #looping for each pivot row
        
        for pivot_row in range(row):
            matrix = Row_operations.partial_pivot(matrix, pivot_row)     #partial pivoting pivot row

            
            #if even after pivoting pivot is zero then no solution
            
            if matrix[pivot_row][pivot_row] == 0:                        
                print("No solution exists")
                return

            else:
                pivot = matrix[pivot_row][pivot_row]
                pivot_inverse = 1/pivot
                
                #multiplying pivot row with inverse of pivot element to make the pivot element 1
                matrix[pivot_row] = Row_operations.multiply_row(matrix, matrix[pivot_row], pivot_inverse)
                
                
                #performing the row addition of each row except the pivot row, with the pivot row, 
                #after multiplying the pivot row with approporiate factor(say s) to make
                #every element in pivot column zero except the pivot element
                #   s = -1*matrix[m][pivot_row]
                
                
                for m in range(row):

                    #looping over each row and skipping pivot row
                    if m != pivot_row:
                        
                        #temp stores a copy of the pivot row multipled by s(defined in prev comments)
                        temp = Row_operations.multiply_row(matrix, matrix[pivot_row], -1*matrix[m][pivot_row])
                        
                        
                        #adding mth row with temp
                        matrix[m] = Row_operations.add_rows(matrix, matrix[m], temp)
        
        
        #rounding off the matrix elements
        matrix = Round.round_matrix(matrix)
        return matrix

    def determinant(matrix):                                #finds the determinant of matrix
        row = len(matrix)                                   #using row operations
        
        for pivot_row in range(row):
            matrix = Row_operations.partial_pivot(matrix, pivot_row)     #partial pivoting pivot row
            
            if matrix[pivot_row][pivot_row] == 0:
                d = 0
                return d
            
            else:
            
                for m in range(row):

                    #performing the row addition of each row except the pivot row, with the pivot row, 
                    #after multiplying the pivot row with approporiate factor(say s) to make
                    #every element in pivot column zero except the pivot element
                    #   s = -1*matrix[m][pivot_row]/matrix[pivot_row][pivot_row]


                    if m != pivot_row:
                        temp = Row_operations.multiply_row(matrix, matrix[pivot_row], -1*(matrix[m][pivot_row])/(matrix[pivot_row][pivot_row]))                  
                        matrix[m] = Row_operations.add_rows(matrix, matrix[m], temp)
                    
        d = 1
        for i in range(row):
            d = d*matrix[i][i]
        
        return d
        
```

```python
##################################################################
'''This class performs all LU operations'''
##################################################################
class LU:
    
    def dolittle(X):
    
        chk = LU.check(X)                                 #checking if LU is possible
        if chk == 1:                                      #check returns 1 if LU not possible
            print("LU decomposition is not possible")
            return

        length = len(X)

        U = Matrix_operations.generate_identity(length)   #making U and L as identity matrices 
        L = Matrix_operations.generate_identity(length)

        for j in range(length):                           #starting iteration
            
            for i in range(j,length):                     #looping to find Ujis
                
                if j == 0:                                #first row in U is always first row in A
                    U[j][i] = X[j][i]
                else:                                     #finding all other elements as per the 
                    summ = 0                              #equations
                    for k in range(j):
                        summ += L[j][k]*U[k][i]                     # (1) 
                    U[j][i] = X[j][i] - summ
                    
                    
            for i in range(j,length):                     #looping to find Lijs
                
                if i != j:                                #if i = j then L[i][j] is anyways 1 because
                                                          #it was identity matrix initially
                    
                    if j == 0:                            #first column in L is first column in A
                        L[i][j] = X[i][j]/U[0][0]         #divided by U[0][0]
                    
                    else:                                 #finding all other elements as per the 
                        summ = 0                          #equations
                        for k in range(j):
                            summ += L[i][k]*U[k][j]
                        L[i][j] = (X[i][j] - summ)/U[j][j]          # (2) 
        return L, U

    def crout(X):

        chk = LU.check(X)
        if chk == 1:
            print("LU decomposition is not possible")
            return

        length = len(X)

        U = Matrix_operations.generate_identity(length)
        L = Matrix_operations.generate_identity(length)
        
        #The difference between codes for crout and dolittle decomposition is that I loop along U and
        #then along L in dolittle but do the reverse in crout because thats how the equations get solved
        #in each case
        
        #The equations are also reversed between L and U. Which can be seen by compairing 1 with 4 and 2
        #with 3

        for j in range(length):
            for i in range(j,length):
                if j == 0:
                    L[i][j] = X[i][j]
                else:
                    summ = 0
                    for k in range(j):
                        summ += L[i][k]*U[k][j]
                    L[i][j] = X[i][j] - summ                      # (3) 
            for i in range(j,length):
                if i != j:
                    if j == 0:
                        U[j][i] = X[j][i]/L[j][j]
                    else:
                        summ = 0
                        for k in range(j):
                            summ += L[j][k]*U[k][i]
                        U[j][i] = (X[j][i] - summ)/L[j][j]        # (4) 

        return L, U
    
    
    def check(X):

        length = len(X)
        chk = 0
            
        for i in range(length):                    #loop to generate submatrices
            Temp = []
            
            for j in range(i+1):
                temp_row = []
                for k in range(i+1):
                    temp_row.append(X[j][k])
                Temp.append(temp_row)
                
            det = Gauss_jordan.determinant(Temp)   #checking determinant of submatrices
            if det == 0:                           #if zero then return
                chk = 1
                return chk
            
        return chk

    
    def solve_dolittle(Z, d):                      #solving equations using dolittle 
        Low, Upp = LU.dolittle(Z)                  #by forward and backward substitution
        y = LU.forward_sub(Low, d)
        x = LU.backward_sub(Upp, y)
        return x

    
    def solve_crout(Z, d):                         #doing the same with crout
        Low, Upp = LU.crout(Z)
        y = LU.forward_sub(Low, d)
        x = LU.backward_sub(Upp, y)
        return x
    
    def solve(L, U, d):
        y = LU.forward_sub(L, d)
        x = LU.backward_sub(U, y)
        return x
    
    def forward_sub(Lower, d):

        length = len(Lower)

        y = []                                     #initialising y = [[0], [0], [0],...]
        for i in range(length):
            temp = []
            temp.append(0)
            y.append(temp)

        y[0][0] = d[0][0]/Lower[0][0]              #finding y_1

        for i in range(1, length):                 #looping to find rest of the elements
            summ = 0                               #as per the equations
            for j in range(i):
                summ += Lower[i][j]*y[j][0]
            y[i][0] = (d[i][0] - summ)/Lower[i][i]

        return y

    
    def backward_sub(Upper, d):

        length = len(Upper)

        x = []                                     #initialising x = [[0], [0], [0],...]
        for i in range(length):
            temp = []
            temp.append(0)
            x.append(temp)

        x[length-1][0] = d[length-1][0]/Upper[length-1][length-1]      #finding x_n

        for i in range(length-2, -1, -1):          #looping to find rest of the elements                      
            summ = 0                               #as per the equations
            for j in range(i+1, length):
                summ += Upper[i][j]*x[j][0]
            x[i][0] = (d[i][0] - summ)/Upper[i][i]

        return x
    
    
    def inverse(X):                              #we will solve using crouts method since either is 
                                                 #fine and serves the same purpose
        length = len(X)   
        
        #creating two copies of X because if I don't do that then X itself gets changed
        
        M = []
        
        for i in range(length):
            temp = []
            for j in range(length):
                temp.append(X[i][j])
            M.append(temp)
            
        K = []
        
        for i in range(length):
            temp = []
            for j in range(length):
                temp.append(X[i][j])
            K.append(temp)
            
        #finding if matrix multiplication is possible
        
        det = Gauss_jordan.determinant(K)
        if det == 0.0:
            print("Inverse doesn't exist")
            return
                    
        #This will become final inverse matrix    
        
        Z = []                                         #initialising Z= [[0], [0], [0],...]
        for i in range(length):
            temp = []
            temp.append(0)
            Z.append(temp)
            
        #starting the loop
        
        for i in range(length):
            
            d = []                                     #initialising d= [[0], [0], [0],...]
            for k in range(length):
                temp = []
                temp.append(0)
                d.append(temp)
                
            #creating a column matrix from each ith column of identity matrix
            
            d[i] = [1]
            
            #augmenting the M and d matrix to initiate partial pivoting in both
            
            aug_mat = Matrix_operations.augment_matrix(M, d)
            
            #partial pivoting each row
            
            for j in range(length):
                aug_mat = Row_operations.partial_pivot(aug_mat, j)
                
            #seperating the augmented matrix to its counterparts
            
            M_new, d_new = Matrix_operations.separate_aug_mat(aug_mat)
            
            #finding the ith columns of the inverse matrix 
            
            temp_col = LU.solve_crout(M_new, d_new)
            
            #appending each column to Z
            
            if i == 0:
                for q in range(length):
                    Z[q] = temp_col[q]
            else:
                for q in range(length):
                    Z[q].append(temp_col[q][0])
                  
        return Z
    
    def cholesky(X):
        n = len(X)

        L = Matrix_operations.generate_identity(n)    #generating L and U as simple identity matrices
        U = Matrix_operations.generate_identity(n)    #they could be anything but for simplicity its identity

        for i in range(n):                            #looping over rows

            for j in range(n):                        #looping over columns

                if i == j:                            #diagonal elements of L

                    summ = 0
                    for k in range(i):
                        summ += L[i][k]**2

                    L[i][j] = math.sqrt(X[i][j] - summ)

                elif i < j:                          #upper triangular elements of L
                    L[i][j] = 0    

                elif i > j:                          #lower triangular elements of L

                    summ = 0
                    for k in range(j):
                        summ += L[i][k]*L[j][k]


                    L[i][j] = (X[i][j] - summ) / L[j][j]

        U = Matrix_operations.transpose(L)           #U is just the transpose of L in case of real entries in X

        return L, U
```

```python
##################################################################
'''This class calculates the roots of functions'''
##################################################################

class Roots:
    def bracket(func, a, b, beta = 1.5, t = 0):                  #bracketing of function around the root
    
        if t <= 12:                                              #terminates if no proper bracketing is found under 12 iterations

            if func(a)*func(b) < 0:                              #condition already satisfied

                return a, b
            elif func(a)*func(b) > 0:                            #condition not satisfied

                t += 1
                if abs(func(a)) <= abs(func(b)):                 #if one value of f less than the other in magnitude, we can
                                                                 #shifting the intervals ends to the left or right depending
                                                                 #on which side of the interval is greater in magnitude
                    a = a - beta*(b - a)
                    return Roots.bracket(func, a, b, beta, t)
                if abs(func(a)) > abs(func(b)):

                    b = b + beta*(b - a)
                    return Roots.bracket(func, a, b, beta, t)
        else:
            return 0                                             #if we get a single digit 0 as returned value we will terminate
                                                                 #the execution and ask user to start again


    def bisection(exp, a, b, eps = 0.00001, graph = False, table = False):   #performs the bisection

        beta = 1.5                                                           #defines the extent to which the interval ends
                                                                             #will shift in bracketing in case initial bracketing
                                                                             #is not valid
        
        if isinstance(exp, str):
            func = lambda x: eval(exp) 
        else:
            func = exp                                          #stores the expression given by the user

        if graph or table:                                                   #when a graph or table is required in the output
            plot_f = []    #list that stores function values                 #to see the trend of the root
            plot_r = []    #list that stores current root values
            plot_j = []    #list that stores iteration values
            j = 1


        if Roots.bracket(func, a, b, beta) == 0:                                   #when initial interval is a bad guess even for bracketing 
            return print("start with a new interval displaced by atleast 15 units to positive or negative direction compared to previous one")
        else:
            a, b = Roots.bracket(func, a, b, beta)


        while abs(b-a) > eps:                                                #runnning the loop until interval size less than
            c = (a + b) / 2                                                  #epsilon

            if func(a) < func(b):                                            #when it is an increasing function
                if func(c) > 0:                                              #if c to the right of actual root, make b = c 
                    b = c
                elif func(c) < 0:                                            #if c to the left of actual root, make a = c 
                    a = c
                else:                                                        #if c is the root, we are done
                    return c
            else:                                                            #when it is a decreasing function
                if func(c) < 0:                                              #if c to the left of actual root, make b = c 
                    b = c
                elif func(c) > 0:                                            #if c to the right of actual root, make a = c 
                    a = c
                else:                                                        #if c is the root, we are done
                    return c 

            if graph or table:                                               #appending the values for plotting
                plot_f.append(func(c))
                plot_r.append(c)
                plot_j.append(j)
                j += 1

        if graph:                                                            #plotting f(x_i) vs i
            plt.plot(plot_j, plot_f)
            plt.title("Convergence of function using bisection")
            plt.xlabel("Number of iternations (i)")
            plt.ylabel("Value of function f(x_i)")
            plt.show()

        if table:                                                            #table
            print("Table of convergence using bisection")
            print("i                x_i                f(x_i)")
            print("__________________________________________")
            for i in range(len(plot_j)):
                print(f"{plot_j[i]}              {round(plot_r[i], 6)}             {round(plot_f[i], 6)}")
                
        return c, plot_f, plot_r, plot_j
    
    def regula_falsi(exp, a, b, eps = 0.00001, graph = False, table = False):
    
    
        beta = 1.5                               #From here to...
        if isinstance(exp, str):
            func = lambda x: eval(exp) 
        else:
            func = exp              
                                                 
        if graph or table:
            plot_f = []
            plot_r = []
            plot_j = []
            j = 1

        if Roots.bracket(func, a, b, beta) == 0:
            return print("start with a new interval displaced by atleast 15 units to positive or negative direction compared to previous one")
        else:
            a, b = Roots.bracket(func, a, b, beta)

                                                #... here the functioning is the same as that for bisection
        condition = True
        c = b - (((b - a) * func(b)) / (func(b) - func(a)))          #setting initial value for c

        while condition:

            if func(a)*func(c) < 0:                                  #if root between a and c
                b = c
                c = b - (((b - a) * func(b)) / (func(b) - func(a)))
                condition = abs(b-c) > eps
            else:                                                    #if root between c and b
                a = c
                c = b - (((b - a) * func(b)) / (func(b) - func(a)))
                condition = abs(a-c) > eps
                
        #Plotting the graph, same as that of bisection

            if graph or table:
                plot_f.append(func(c))
                plot_r.append(c)
                plot_j.append(j)
                j += 1

        if graph:
            plt.plot(plot_j, plot_f)
            plt.title("Convergence of function using regula falsi")
            plt.xlabel("Number of iternations (i)")
            plt.ylabel("Value of function f(x_i)")
            plt.show()

        if table:
            print("Table of convergence using regula falsi")
            print("i                x_i                f(x_i)")
            print("__________________________________________")
            for i in range(len(plot_j)):
                print(f"{plot_j[i]}         {round(plot_r[i], 6)}         {round(plot_f[i], 6)}")

        return c, plot_f, plot_r, plot_j
    
    def derivative(func, p, h = 0.00001):                     #calculating the derivative of a function using standard formula
    
        d = (func(p + h) - func(p - h)) / (2 * h)

        return d

    def newton_raphson(exp, k, eps = 0.00001):                #using newton raphson to find the roots
        
        if isinstance(exp, str):
            func = lambda x: eval(exp) 
        else:
            func = exp
        
        root = k - (func(k) / Roots.derivative(func, k))            #calculating root from initial guess 

        condition = True

        while condition:

            root_1 = root - (func(root) / Roots.derivative(func, root))   #refining the root

            condition = abs(root_1 - root) > eps                    #condition is false when root is achieved at desired accuracy

            root = root_1                                           #setting root as the refined root to refine it further in
                                                                    #in the next iteration
        return root
    
    def deflate(coeff, q):                                          #deflating a polynomial P(x) to (x-q)Q(x) where q is a root of P(x)
    
        final = [coeff[0]]

        for i in range(1, len(coeff)):                              #here we have used synthetic division algorithm
            final.append((q * final[i-1]) + coeff[i])

        final.remove(final[len(final) - 1])

        return final                                                #return the coefficient matrix of Q(x)

    def laguerre(coeff = [], x0 = 0, eps = 0.00001):                #calculating all roots of a polynomial P(x)(restricted to real roots)
                                                                    #x0 is a known root
        root = []
        n = len(coeff) - 1                                          #n is degree of polynomial
        h = 0.000001

        while n > 1:                                                #iterates to find roots of P(x) by finding one root
                                                                    #of every deflated polynomial of P(x) until a one degree
                                                                    #polynomial is reached
                    
            func = lambda x: sum([(coeff[i])*x**(n-i) for i in range(n+1)])   #generating polynomial function from coefficient matrix

            while True:                                             #loop terminates when the roots are close to the desired accuracy
                if abs(func(x0)) < eps:
                    coeff = Roots.deflate(coeff, x0)                  #when one root is found, polynomial is deflated by one degree and 
                                                                    #next root is found in next loop
                        
                    break                                           #breaks from the if else checking
                    
                else:                                               #when the guessed root needs refining
                    
                    
                    #equations to refine the root by laguerre method
                    G = Roots.derivative(func, x0) / func(x0)
                    H = G*G - ((Roots.derivative(func, x0 + h) - Roots.derivative(func, x0 - h)) / (2 * h * func(x0)))
                    
                    #selecting the larger of the denominators for better refining
                    if G + math.sqrt((n-1)*(n*H - G*G)) > G - math.sqrt((n-1)*(n*H - G*G)):
                        a = n / (G + math.sqrt((n-1)*(n*H - G*G)))
                    else:
                        a = n / (G - math.sqrt((n-1)*(n*H - G*G)))
                        
                    x0 = x0 - a                                     #setting the refined root
            #appending 
            root.append(x0)                                        #appending the currently found root of P(x) to a list
            n = n - 1                                               #reducing degree by one to find next root in the next loop

        root.append(-coeff[1] / coeff[0])                          #the last root for a one degree polynomial needs special
                                                                    #code in the end because its deflation is not possible

        return root                                                #returns the roots
```

```python
##################################################################
'''This class calculates the integrals of functions'''
##################################################################

class Integral:
    
    def midpoint(exp, a, b, N = 10, err = 0, fmax = 0):         
    
        if isinstance(exp, str):
            func = lambda x: eval(exp) 
        else:
            func = exp                    #defining the function and h
            
        summ = 0
        temp = a          #temp will be a first, then a+h, then a+2h and so on in the loop

        if err == 0:                                   #err = 0 implies that we will use given value of N
            h = (b-a)/N
        else:                                          #err != 0 implies we will find N
            N = pow((((b-a)**3)*fmax)/(err*24), 1/2)
            N = math.ceil(N)
            h = (b-a)/N

        for i in range(N):                             #calculating the integral
            summ += func((2*temp + h)/2)
            temp = temp + h

        return h*summ
    
    def trapezoid(exp, a, b, N = 0, err = 0, fmax = 0):         #similar calculations as midpoint with different formulas
    
        if isinstance(exp, str):
            func = lambda x: eval(exp) 
        else:
            func = exp
        summ = 0
        h = (b-a)/N
        temp = a

        if err == 0:
            h = (b-a)/N
        else:
            N = pow((((b-a)**3)*fmax)/(err*12), 1/2)
            N = math.ceil(N)
            h = (b-a)/N

        for i in range(N):                               #using trapezoid method for loop
            summ += (func(temp)+func(temp+h))
            temp = temp + h

        return h*0.5*summ
    
    def simpson(exp, a, b, N = 10, err = 0, fmax = 0):           #similar calculation like in midpoint till h is defined
    
        if isinstance(exp, str):
            func = lambda x: eval(exp) 
        else:
            func = exp
        summ = 0
        temp = a

        if err == 0:
            h = (b-a)/N
        else:
            N = pow(((((b-a)**5)*fmax)/(err*180)), 1/4)
            N = math.ceil(N)
            h = (b-a)/N

        for i in range(N+1):                                    #summing over the areas with weight function as defined for simpsons method
            if i == 0 or i == N:
                summ += func(temp)
            elif i>0 and i<N and i%2 == 0:
                summ += 2*func(temp)
            elif i>0 and i<N and i%2 != 0:
                summ += 4*func(temp)

            temp = temp + h

        summ = h*summ/3                           #multiplying h/3 factor to get the integral
        return summ
    
    def monte_carlo_step(exp, a, b, rng, N = 10, step = 10, pt = False, mult = False):
    
        if isinstance(exp, str):
            func = lambda x: eval(exp) 
        else:
            func = exp

        F = []            #stores F_n
        S = []            #stores sigma_n
        plot_n = []

        while N < rng:

            summ = 0                   #summ is F_n
            sqsumm = 0                 #sqsumm is sigma_n
            h = abs(b - a) / N         #constant factor to be multiplied to summ later

            for i in range(N):         #looping to calculate F_n and sigma_n
                x = rand.uniform(a, b)
                summ += func(x)
                sqsumm += func(x)*func(x)
                
            summ = h*summ
            sqsumm = (h*sqsumm) - summ*summ                  #calculating sigma_n
                                                 #multiplying constant factor to summ

            F.append(summ)
            S.append(sqsumm)
            plot_n.append(N)
            
            if mult == False:
                N += step                  #incrementing N by 10 for each iteration
            else:
                N += step*N
        if pt == True:                   #plotting the convergence of roots
            plt.plot(plot_n, F)
            plt.title("Convergence of the integral with N")
            plt.xlabel("N")
            plt.ylabel("Value of integral")
            plt.show()

        return F, S
    
    def monte_carlo(exp, a, b, N = 10, pt = False):
    
        if isinstance(exp, str):
            func = lambda x: eval(exp) 
        else:
            func = exp


        summ = 0                   #summ is F_n
        sqsumm = 0                 #sqsumm is sigma_n
        h = abs(b - a) / N         #constant factor to be multiplied to summ later

        for i in range(N):         #looping to calculate F_n and sigma_n
            x = rand.uniform(a, b)
            summ += func(x)
            sqsumm += func(x)*func(x)
            
        var = (sqsumm/N) - (summ/N)**2      #calculating sigma_n
        summ = h*summ                                     #multiplying constant factor to summ

        if pt == True:                   #plotting the convergence of roots
            plt.plot(plot_n, F)
            plt.title("Convergence of the integral with N")
            plt.xlabel("N")
            plt.ylabel("Value of integral")
            plt.show()

        return summ, var
```

```python
class Deq:
    
    def euler_forward(func, xin, yin, xfin, h = 0.00001, plot = False):       #y(x+h) = y(x) + hf(y(x),x)
    
        y = yin
        x = xin
        X = [xin]
        Y = [yin]

        while x < xfin:                                      #looping till whatever final value we desire

            y += h*func(x, y)
            x += h  
            Y.append(y)
            X.append(x)

        if plot == True:

            title = "y(x) found when h = " + str(h)

            plt.plot(X, Y)
            plt.title(title)
            plt.xlabel("x")
            plt.ylabel("y")
            plt.show()

        else:
            
            return X, Y
    
    
    def pred_corr(func, xin, yin, xfin, h = 0.00001, plot = False):       #y(x+h) = y(x) + hf(y(x),x)
    
        y = yin
        x = xin
        X = [xin]
        Y = [yin]

        while x < xfin:                                      #looping till whatever final value we desire

            yp = y + h*func(x, y)
            y += h*0.5*(func(x, y) + func(x + h, yp))
            x += h  
            Y.append(y)
            X.append(x)

        if plot == True:

            title = "y(x) found when h = " + str(h)

            plt.plot(X, Y)
            plt.title(title)
            plt.xlabel("x")
            plt.ylabel("y")
            plt.show()

        else:
            
            return X, Y
    
    
    def RK3_ON(X, t, tf, F, n, h = 0.001):       
    
    #same as predictor corrector
    #takes X as matrix of all initial values of variables. Diff equation is of the form dX[i]/dt = f(X[1], X[2],..., X[n], t)
    
    
        T = [t]                         #matrix that stores all t values
        Var = []                        #matrix that will store the progression of all the variables
        for i in range(n):
            Var.append([X[i]])

        while t < tf:

            K1 = []                     #temporary storage of K1, K2 and K3 values
            K2 = []
            K3 = []
            xtemp = []                  #temporary storage of X[i] + K1/2 (or K2 or K3)so that it can be used when finding the value of the function

            for i in range(n):
                K1.append(h*F[i](X, t))              #K1 stores all values of k1 for all the variables
            for i in range(n):
                xtemp.append(X[i] + K1[i]/0.5)       
            for i in range(n):
                K2.append(h*F[i](xtemp, t + h/2))    #K2 stores all values of k2 for all the variables
            xtemp = []
            for i in range(n):
                xtemp.append(X[i] + 2*K2[i] - K1[i])     
            for i in range(n):
                K3.append(h*F[i](xtemp, t + h))      #K3 stores all values of k3 for all the variables
            for i in range(n): 
                X[i] += (K1[i] + 4*K2[i] + K3[i])/6  #finding X[i], i.e. values of all the variables for some t
                Var[i].append(X[i])

            t += h                             #increasing t by a small amount
            T.append(t)

        return Var, T
    
    def RK4_O2(x, y, t, tf, fx, fy, h = 0.01):   #x and y are the variables
    
        T = [t]
        X = [x]
        Y = [y]

        while t < tf:

            kx1 = h*fx(x, y, t)
            ky1 = h*fy(x, y, t)

            kx2 = h*fx(x + kx1/2, y + ky1/2, t + h/2)
            ky2 = h*fy(x + kx1/2, y + ky1/2, t + h/2)

            kx3 = h*fx(x + kx2/2, y + ky2/2, t + h/2)
            ky3 = h*fy(x + kx2/2, y + ky2/2, t + h/2)

            kx4 = h*fx(x + kx3, y + ky3, t + h)
            ky4 = h*fy(x + kx3, y + ky3, t + h)

            t += h
            x += (kx1 + 2*(kx2 + kx3) + kx4)/6
            y += (ky1 + 2*(ky2 + ky3) + ky4)/6

            T.append(t)
            X.append(x)
            Y.append(y)

        return X, Y, T
    
    
    #this is an RK4 method that can solve N ordered differential equation. The method is same as RK3, just different formula. Please check above.
    def RK4_ON(X, t, tf, func, n, h = 0.01):
    
        T = [t]
        Var = []
        for i in range(n):
            Var.append([X[i]])
            
        if t < tf:
            while t < tf:

                K1 = []
                K2 = []
                K3 = []
                K4 = []
                xtemp = []

                for i in range(n):
                    K1.append(h*func[i](X, t))
                for i in range(n):
                    xtemp.append(X[i] + K1[i]/0.5) 
                for i in range(n):
                    K2.append(h*func[i](xtemp, t + h/2))
                xtemp = []
                for i in range(n):
                    xtemp.append(X[i] + K2[i]/0.5)     
                for i in range(n):
                    K3.append(h*func[i](xtemp, t + h/2))
                xtemp = []
                for i in range(n):
                    xtemp.append(X[i] + K3[i]) 
                for i in range(n):
                    K4.append(h*func[i](xtemp, t + h))

                for i in range(n): 
                    X[i] += (K1[i] + 2*(K2[i] + K3[i]) + K4[i])/6
                    Var[i].append(X[i])

                t += h
                T.append(t)
                
        else:
            while t > tf:
                
                K1 = []
                K2 = []
                K3 = []
                K4 = []
                xtemp = []

                for i in range(n):
                    K1.append(h*func[i](X, t))
                for i in range(n):
                    xtemp.append(X[i] + K1[i]/0.5) 
                for i in range(n):
                    K2.append(h*func[i](xtemp, t + h/2))
                xtemp = []
                for i in range(n):
                    xtemp.append(X[i] + K2[i]/0.5)     
                for i in range(n):
                    K3.append(h*func[i](xtemp, t + h/2))
                xtemp = []
                for i in range(n):
                    xtemp.append(X[i] + K3[i]) 
                for i in range(n):
                    K4.append(h*func[i](xtemp, t + h))

                for i in range(n): 
                    X[i] += (K1[i] + 2*(K2[i] + K3[i]) + K4[i])/6
                    Var[i].append(X[i])

                t += h
                T.append(t)
                
        return Var, T
    
    def RK2_ON(X, t, tf, func, n, h = 0.01):

        T = [t]
        Var = []
        for i in range(n):
            Var.append([X[i]])

        while t < tf:

            K1 = []
            K2 = []
            K3 = []
            K4 = []
            xtemp = []

            for i in range(n):
                K1.append(h*func[i](X, t))
            for i in range(n):
                xtemp.append(X[i] + K1[i]/0.5) 
            for i in range(n):
                K2.append(h*func[i](xtemp, t + h/2))

            for i in range(n): 
                X[i] += K2[i]
                Var[i].append(X[i])

            t += h
            T.append(t)

        return Var, T
```

```python
class Shootingclass:
    #xi = initial boundary condition of X[0] at t = i
    #xf = final boundary condition of X[0] at t = f
    #yi_h = first manual guess of X[1] at t = i
    #yi_l = correction to the guess yi_h
    #F = list of finctions
    #error = allowed error
    #tf = final range of time while calculating for RK4

    def check_shoot(xi, xf, yi_l, i, f, F, error, h = 0.1, epsilon = 0.01):

        X = [xi, yi_l]
        V, T = Deq.RK4_ON(X, i, f, F, 2, h)

        if abs(V[0][-1] - xf) >= error:
            return False
        else:
            return(None)

    def shooting(xi, xf, yi_h, i, f, F, h = 0.1, err = 0.1,  epsilon = 0.01):

        X = [xi, yi_h]
        V, T = Deq.RK4_ON(X, i, f, F, 2, h)


        length = len(T)


        xf_h = V[0][-1]                     #xf_h = X[0] at t = f for the solutions found above using RK4

        if abs(V[0][-1] - xf) <= err:
            return V, T

        while abs(V[0][-1] - xf) > err:     #while X[0][final], i.e. X[0] for t = f as given by RK4, is within err distance from true X[0] at t = f as given in the initial conditions
            error_true = abs(V[0][-1] - xf)

            if Shootingclass.check_shoot(xi, xf, yi_h + 0.2, i, f, F, error_true, h) == False:   #check if changing initial guess of X[1] at t = i by some higher amount is giving less error
                yi_l = yi_h - epsilon                                              #if not then subtract and label new initial conditon with an l
                X = [xi, yi_l]                                                     #changing X by modifying the better initial condition of X[1]
                V, T = Deq.RK4_ON(X, i, f, F, 2, h)    #perform RK4
                yi_h = yi_l

            else:
                yi_l = yi_h + epsilon
                X = [xi, yi_l]                     #changing X by modifying the better initial condition of X[1]
                V, T = Deq.RK4_ON(X, i, f, F, 2, h)   #perform RK4
                yi_h = yi_l


        xf_l = V[0][-1]                     #xf_l is the corrected and closer to actual value of X[0] at t = f, AS PER RK4

        yi = yi_l + (yi_h - yi_l)*(xf - xf_l)/(xf_h - xf_l)    #use l and h labeled initial conditions of X[1] to get a modified one

        if Shootingclass.check_shoot(xi, xf, yi, i, f, F, err, h) == False:   #check if this is close enough to the boundary conditions
            Shootingclass.shooting(xi, xf, yi, i, f, F, h, err)    #if not then do it again with improved value of yi

        else:

            return V, T
        
    #xi = initial boundary condition of X[0] at t = i
    #xf = final boundary condition of X[0] at t = f
    #yi_h = first manual guess of X[1] at t = i
    #yi_l = correction to the guess yi_h
    #F = list of finctions
    #error = allowed error
    #tf = final range of time while calculating for RK4

    def check_shoot1(xi, xf, yi_l, i, f, F, error, c = 'g', h = 0.1):

        X = [xi, yi_l]
        V, T = Deq.RK4_ON(X, i, f, F, 2, h)

        if c == 'g':
            if V[0][-1] - xf < 0 or abs(V[0][-1] - xf) < error:
                return True
            else:
                False

        else:
            if V[0][-1] - xf > 0 or abs(V[0][-1] - xf) < error:
                return True
            else:
                False

    def shooting1(xi, xf, yi_h, i, f, F, h = 0.1, err = 0.1,  epsilon = 0.01):

        X = [xi, yi_h]
        V, T = Deq.RK4_ON(X, i, f, F, 2, h)
        length = len(T)


        xf_h = V[0][-1]                     #xf_h = X[0] at t = f for the solutions found above using RK4
        yi_t = yi_h

        if abs(V[0][-1] - xf) <= err:
            return V, T

        while abs(V[0][-1] - xf) > err:     #while X[0][final], i.e. X[0] for t = f as given by RK4, is within err distance from true X[0] at t = f as given in the initial conditions

            if V[0][-1] - xf > 0:

                yi_l = yi_t - epsilon
                if Shootingclass.check_shoot1(xi, xf, yi_l, i, f, F, V[0][-1] - xf, 'g', h) == False:
                    yi_l = yi_t + 2*epsilon
                    yi_t = yi_l
                else:
                    yi_t = yi_l

                X = [xi, yi_l]
                V, T = Deq.RK4_ON(X, i, f, F, 2, h)

            else:

                yi_l = yi_t - epsilon
                if Shootingclass.check_shoot1(xi, xf, yi_l, i, f, F, V[0][-1] - xf, 'l', h) == False:
                    yi_l = yi_t + 2*epsilon
                    yi_t = yi_l
                else:
                    yi_t = yi_l


                X = [xi, yi_l]
                V, T = Deq.RK4_ON(X, i, f, F, 2, h)


        xf_l = V[0][-1]                                         #xf_l is the corrected and closer to actual value of X[0] at t = f, AS PER RK4
        yi = yi_l + (yi_h - yi_l)*(xf - xf_l)/(xf_h - xf_l)     #use l and h labeled initial conditions of X[1] to get a modified one

        if Shootingclass.check_shoot1(xi, xf, yi, i, f, F, err, h) == False:   #check if this is close enough to the boundary conditions

            Shootingclass.shooting(xi, xf, yi, i, f, F, h, err)               #if not then do it again with improved value of yi

        else:
            return V, T
    
```
